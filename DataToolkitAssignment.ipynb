**Data Toolkit Questions**

1. What is NumPy, and why is it widely used in Python?
>> NumPy also Numerical Python is library used for data analysis and computation. It is widely used to create multidimensional arrays and matrices and it also contains multiple functions to operate on these arrays.

2. How does broadcasting work in NumPy?
>> Broadcasting in NumPy allows arithmetic operations on matrixes or array of different shape by automatically expanding smaller array to match the size of larger one.

3. What is a Pandas DataFrame?
>> Pandas dataframe is a two dimensional data structutre with rows and columns which is used to store heterogeneous data and supports high level mathematical functions and computations on the rows and columns

4. Explain the use of the `groupby()` method in Pandas.
>> Groupby() in pandas is a method used to apply certain computations or operations on a set of cloumns by dividing them into different groups. Suppose we have a data containing number of students in different class and there sex. We can use groupby to give count of total students by sex in all the class or by classwise.

5. Why is Seaborn preferred for statistical visualizations?
>> Seaborn built on top of matplotlib supports aesthetitcally pleasing and informative visualization for a variety of plots for all sorts of data. Hence it is widely used for statistical visualizations.

6. What are the differences between NumPy arrays and Python lists?
>> Numpy array is a homogeneous data structure means it can only contain data of same type which is widely used to store data and perfrom statistical operations. While Lists in python supports all sort of data types and is generally used for looping.

7. What is a heatmap, and when should it be used?
>> A heatmap is a graphical representation of data where individual values are represented using different colors. It is commonly used to visualize correlation matrices, feature importance, and data density distributions.

8. What does the term “vectorized operation” mean in NumPy?
>> Vectorized operation in NumPy means performing operations on rows/arrays instead of individual elements.

9. How does Matplotlib differ from Plotly?
>> Matplotlib is a static visualization library, primarily used for 2D plotting with extensive customization options. Plotly is an interactive visualization library, allowing zooming, hovering, and real-time updates, making it ideal for dashboards.

10. What is the significance of hierarchical indexing in Pandas?
>>  Hierarchical Indexing also Multi Indexing allows working with a two dimensional structure cotaining rows and column labels. It can be used to represent nested data efficiently.

11. What is the role of Seaborn’s pairplot() function?
>> The pairplot() function creates pairwise scatter plots of numerical columns in a dataset. It is useful for exploring relationships between variables, detecting correlations, and visualizing distributions.

12. What is the purpose of the describe() function in Pandas?
>> Describe() function in Pandas gives you a statistical summary of data frame with important measures such as mean, median, max, min and standar deviation.

13. Why is handling missing data important in Pandas?
>> Missing data can lead to incorrect analysis and errors in machine learning models. Pandas provides methods like dropna(), fillna(), and interpolate() to handle missing values effectively.

14. What are the benefits of using Plotly for data visualization?
>> Plotly can be used to create interactive and real time dasboards and charts with support of wide variety charts including 3D visualizations.

15. How does NumPy handle multidimensional arrays?
>> NumPy uses the ndarray object to store multidimensional arrays efficiently. It allows indexing, slicing, and broadcasting across multiple dimensions, enabling complex mathematical operations on high-dimensional data.

16. What is the role of Bokeh in data visualization?
>> Bokeh is a Python library for creating interactive visualizations for web applications. It is particularly useful for large datasets and integrates well with Jupyter Notebooks and web frameworks.

17. Explain the difference between apply() and map() in Pandas.
>> apply() function can be used to apply any function across a column or row in a dataframe.  map() is used to apply a function on elements of a series instead of applying on dataframes directly.

18. What are some advanced features of NumPy?
>> Some advanced features of NumPy are: a. Broadcasting: Automatically expanding the dimesnion of an array for operations between arrays of different shapes, b. Random sampling and distributions using random method, c. Linear algebra functions for advanced mathematical and statistical operations. d. Advanced indexing making slicing easier and faster.

19. How does Pandas simplify time series analysis?
>> Pandas provides powerful tools for working with time series data, such as date-based indexing, resampling, and rolling-window calculations, making it easier to analyze trends over time.

20. What is the role of a pivot table in Pandas?
>> Pivot table in Pandas is similar to pivot table in excel allowing to reshape a data frame into desired columns and rows and applying an aggregate function summarizing relevant data.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
>> NumPy arrays are stored in contiguous memory locations, allowing efficient access and manipulation using optimized C-based operations. Python lists, on the other hand, store elements as references, making slicing slower

22. What are some common use cases for Seaborn?
>> Seaborn is used for advanced statical visualization. Common use cases are summarizing a data using boxplots or bar graphs. Showing relationships between variables using scatter plots. Generating heat maps and violin plots for analysis.

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


df = sns.load_dataset('iris')

corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')


plt.title('Correlation Matrix Heatmap')

plt.show()
```

```python
#5. Generate a bar plot using Plotly.
import plotly.express as px
import pandas as pd

data = {
    'Name': ['Shubham', 'Ashni', 'Kiran', 'Shiwangi','Sonia','Om'],
    'Marks': [94, 75, 80, 56, 78,90]
}

df = pd.DataFrame(data)

fig = px.bar(df, x='Name', y='Marks', title='Marks Obtained by Students')

fig.show()
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays.

ar1=np.ones((3,3))
ar2=np.array([[3,6,7],[5,2,5],[9,2,4]])
print("Ar1:\n",ar1)
print("Ar2:\n",ar2)
print("Element wise multiplication of Ar1 and Ar2:")
print(ar1*ar2)
```

```python
#8. Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 12, 14, 16, 18]
y2 = [8, 11, 13, 17, 20]
y3 = [5, 7, 9, 12, 15]

plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='green', marker='s')
plt.plot(x, y3, label='Line 3', color='red', marker='^')

plt.title('Multiple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.grid(True)

plt.show()
```

```python
#13. Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame({
    'X': np.random.randn(100),
    'Y': np.random.randn(100),
    'Z': np.random.randn(100),
    'Category': np.random.choice(['A', 'B'], size=100)
})

fig = px.scatter_3d(df, x='X', y='Y', z='Z',
                    color='Category',
                    title='3D Scatter Plot Example')
fig.show()
```

4. How does loss value help in determining whether the model is good or not?
>>  Loss values help find out how much the preditictions varies from the actual outcomes. Thus if loss value is high then the model is not good, lower loss values means model is more robust.

5. What are continuous and categorical variables?
>> Continuous variables are numeric values which can take any value on a scale. Example are height, distance. Categorical variables are non numeric variables such as gender, grade or color.

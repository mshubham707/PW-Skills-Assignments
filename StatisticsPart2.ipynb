# **Statistics Part 2**

1. What is hypothesis testing in statistics?
>> Hypothesis testing is a statistical method to make decisions using data. It tests an assumption (hypothesis) about a population parameter.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
>> The null hypothesis (H₀) assumes no effect or difference, while the alternative hypothesis (H₁ or Ha) suggests a significant effect or difference.

3. What is the significance level in hypothesis testing, and why is it important?
>> The significance level (α) is the probability of rejecting the null hypothesis when it's actually true (Type I error). Common values: 0.05, 0.01.

4. What does a P-value represent in hypothesis testing?
>> A P-value is the probability of obtaining test results at least as extreme as the observed ones, assuming the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?
>> If P ≤ α: Reject H₀ : evidence supports H₁

 >> If P > α: Fail to reject H₀ : insufficient evidence for H₁

6. What are Type 1 and Type 2 errors in hypothesis testing?
>> Type I error: Rejecting a true null hypothesis. Also known as False Positive

 >> Type II error: Failing to reject a false null hypothesis. Also known as False Negative

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
>> One-tailed test: Tests for effect in one direction

 >> Two-tailed test: Tests for effect in both directions

8. What is the Z-test, and when is it used in hypothesis testing?
>> Z-test is a hypothesis testing method used in statistics to find out whether a finding is statistically significant or not. Z-test is used when the population variance is known and the sample size is large (n > 30).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

>> Z = (X̄ - μ) / (σ/√n)
It measures how many standard deviations a sample mean is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
>> T-distribution is used when the sample size is small(<30) and population variance is not known.

11. What is the difference between a Z-test and a T-test?
>> Z-test uses known population variance while T-test uses sample standard deviation and is for small samples.


12. What is the T-test, and how is it used in hypothesis testing?
>> A T-test checks if there is a significant difference between means (one sample, two samples, or paired samples). It is used similarly like a z test in its place when there is unknown population variance and sample size is less than 30.

13. What is the relationship between Z-test and T-test in hypothesis testing?
>> Both test hypotheses about means. T-test is used when the population SD is unknown or sample is small; Z-test otherwise.

14. What is a confidence interval, and how is it used to interpret statistical results?
>> A confidence interval gives a range of values within which the population parameter is expected to lie with a certain confidence (e.g., 95%).

15. What is the margin of error, and how does it affect the confidence interval?
>> Margin of error is measured by dividing sample standard deviation by square root of sample size and multiplying it with the statistic. It measures the range of uncertainity around an estimate.

16. How is Bayes' Theorem used in statistics, and what is its significance?
>> Bayes' Theorem is a foundational concept in statistics, especially in the field of inferential statistics and probability theory. It describes how to update the probability of a hypothesis based on new evidence.

17. What is the Chi-square distribution, and when is it used?
>> Chi square distribution is obtained by summation of squares of a standard normal random variate. It's used for categorical data to test independence or goodness of fit.

18. What is the Chi-square goodness of fit test, and how is it applied?
>> It compares the observed frequencies to an expected frequency and evaluate whether there exists some relationship or not. It is applied by calculating chi square statistic and comparing it with the required p value.

19. What is the F-distribution, and when is it used in hypothesis testing?
>> F distribution is used to compare variances across groups and to compare multiple means within and among groups. It is generally used in ANOVA testing.

20. What is an ANOVA test, and what are its assumptions?
>> ANOVA (Analysis of Variance) compares means of 3+ groups. Assumes independence, normality, and equal variances.

21. What are the different types of ANOVA tests?
>> One-way ANOVA

 >> Two-way ANOVA

 >> Repeated Measures ANOVA

22. What is the F-test, and how does it relate to hypothesis testing?
>> The F-test is a statistical test that compares the variances of two or more populations. It's primarily used in hypothesis testing to determine if there's a significant difference between the means of these populations. If the variances are significantly different, it suggests that there's also a significant difference in the means, thus rejecting the null hypothesis

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import pandas as pd
from scipy.stats import norm

P_data=sns.load_dataset("tips")["tip"]
p_mean=P_data.mean().round(3)

sample_tip=P_data.sample(n=40)
s_size=40
s_mean=sample_tip.mean()
s_std=np.std(sample_tip,ddof=1)

def hypo_test(pop_mean,samp_mean,samp_std,s_size,alp=0.05):
  print(f"Null Hypothesis: mu = {pop_mean}")
  print(f"Alt. Hypothesis: mu != {pop_mean}")
  z_score=(samp_mean-pop_mean)/(s_std/np.sqrt(s_size))
  z_critical=norm.ppf(1-alp/2)
  print(f"Z-critical: {z_critical:.3f}")
  print(f"Z-score: {z_score:.3f}")
  if z_critical < z_score or -z_critical >z_score:
    return "Reject null hypothesis"
  else:
    return "Failed to reject null hypothesis"

hypo_test(p_mean,s_mean,s_std,s_size)


```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mean=0, std=1, n_samples=1000):

    data = np.random.normal(mean, std, n_samples)
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)

    return data_skewness, data_kurtosis


skewness_val, kurtosis_val = calculate_skewness_kurtosis()
print(f"Skewness: {skewness_val:.4f}")
print(f"Kurtosis: {kurtosis_val:.4f}")
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import norm

np.random.seed(2)
P_data=sns.load_dataset("tips")["tip"]
p_mean=P_data.mean().round(3)

sample_tip=P_data.sample(n=40)
s_size=40
s_mean=sample_tip.mean()
s_std=np.std(sample_tip,ddof=1)

def hypo_test(pop_mean,samp_mean,samp_std,s_size,alp=0.05):
  print(f"Null Hypothesis: mu = {pop_mean}")
  print(f"Alt. Hypothesis: mu > {pop_mean}")
  z_score=(samp_mean-pop_mean)/(s_std/np.sqrt(s_size))
  z_critical=norm.ppf(1-alp)
  print(f"Z-critical: {z_critical:.3f}")
  print(f"Z-score: {z_score:.3f}")
  if z_score > z_critical:
    return "Reject null hypothesis"
  else:
    return "Failed to reject null hypothesis"


result=hypo_test(p_mean,s_mean,s_std,s_size)
print(result)

print("\nInterpretation:")
if result == "Reject null hypothesis":
    print(f"There is statistically significant evidence to conclude that the sample mean is greater than the population mean of {p_mean}.")
else:
    print(f"There is no statistically significant evidence to conclude that the sample mean is greater than the population mean of {p_mean}.")
```

```python
#2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

import numpy as np
from scipy.stats import norm


np.random.seed(40)
pop = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10, n=50

pop_mean = 50
sample=np.random.choice(pop,size=40,replace=True)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)


z_score = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))


p_value = 2 * (1 - norm.cdf(abs(z_score)))

alpha = 0.05

print(f"Sample Mean: {sample_mean:.3f}")
print(f"Sample Standard Deviation: {sample_std:.3f}")
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance level: {alpha}")

if p_value < alpha:
    print("Reject the null hypothesis: significant difference found.")
else:
    print("Fail to reject the null hypothesis: no significant difference found.")

```

```python
#3.  Implement a one-tailed Z-test using Python to compare the sample mean with the population mean.

import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import norm

np.random.seed(33)
P_data=sns.load_dataset("tips")["tip"]
p_mean=P_data.mean().round(3)

sample_tip=P_data.sample(n=40)
s_size=40
s_mean=sample_tip.mean()
s_std=np.std(sample_tip,ddof=1)

alp=0.05

print(f"Null Hypothesis: mu = {p_mean}")
print(f"Alt. Hypothesis: mu < {p_mean}")
z_score=(s_mean-p_mean)/(s_std/np.sqrt(s_size))

z_critical=norm.ppf(alp)

print(f"Z-critical: {z_critical:.3f}")
print(f"Z-score: {z_score:.3f}")

if z_score < z_critical:
  print("Reject the Null hypothesis")
else:
  print("Failed to reject the null hypothesis")


```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(33)
pop = np.random.normal(loc=50, scale=10, size=1000)  # mean=52, std=10, n=50

pop_mean = 50
sample=np.random.choice(pop,size=40,replace=True)
s_mean = np.mean(sample)
s_std = np.std(sample, ddof=1)
n = len(sample)

alp=0.05

print(f"Null Hypothesis: mu = {pop_mean}")
print(f"Alt. Hypothesis: mu != {pop_mean}")
print(f"Sample Mean: {s_mean:.2f}")
print(f"Sample std: {s_std:.2f}")
z_score=(s_mean-pop_mean)/(s_std/np.sqrt(n))

z_critical=norm.ppf(1-alp/2)

print(f"Z-critical: {z_critical:.3f}")
print(f"Z-score: {z_score:.3f}")

if abs(z_score) > z_critical:
  print("Reject the Null hypothesis")
else:
  print("Failed to reject the null hypothesis")

print("\n")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

#plotting the decision region
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')


plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region')


plt.axvline(z_score, color='black', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_critical, color='red', linestyle=':', label=f'Z = ±{z_critical:.2f}')
plt.axvline(z_critical, color='red', linestyle=':')

plt.title("Two-tailed Z-test Decision Region")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0=50, mu1=52, sigma=10, n=30, alpha=0.05, test_type='right'):
    """
    Visualizes Type I and Type II errors in hypothesis testing.

    Parameters:
    - mu0: Mean under null hypothesis (H0)
    - mu1: Mean under alternative hypothesis (H1)
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    - test_type: 'right', 'left', or 'two' for one- or two-tailed test
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical values
    if test_type == 'right':
        z_crit = norm.ppf(1 - alpha)
        crit_value = mu0 + z_crit * se
        beta = norm.cdf((crit_value - mu1) / se)
    elif test_type == 'left':
        z_crit = norm.ppf(alpha)
        crit_value = mu0 + z_crit * se
        beta = 1 - norm.cdf((crit_value - mu1) / se)
    elif test_type == 'two':
        z_crit = norm.ppf(1 - alpha / 2)
        upper_crit = mu0 + z_crit * se
        lower_crit = mu0 - z_crit * se
        beta = norm.cdf((upper_crit - mu1) / se) - norm.cdf((lower_crit - mu1) / se)

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label=f'H0: μ={mu0}', color='blue')
    plt.plot(x, y1, label=f'H1: μ={mu1}', color='green')

    if test_type == 'right':
        plt.fill_between(x, y0, where=(x > crit_value), color='red', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x, y1, where=(x < crit_value), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(crit_value, color='black', linestyle='--', label=f'Critical value = {crit_value:.2f}')
    elif test_type == 'left':
        plt.fill_between(x, y0, where=(x < crit_value), color='red', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x, y1, where=(x > crit_value), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(crit_value, color='black', linestyle='--', label=f'Critical value = {crit_value:.2f}')
    elif test_type == 'two':
        plt.fill_between(x, y0, where=(x > upper_crit) | (x < lower_crit), color='red', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x, y1, where=(x > lower_crit) & (x < upper_crit), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(upper_crit, color='black', linestyle='--', label=f'Upper crit = {upper_crit:.2f}')
        plt.axvline(lower_crit, color='black', linestyle='--', label=f'Lower crit = {lower_crit:.2f}')

    plt.title("Type I and Type II Error Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

plot_type1_type2_errors()
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind


np.random.seed(42)

group1 = np.random.normal(loc=100, scale=10, size=50)
group2 = np.random.normal(loc=105, scale=10, size=50)


t_stat, p_value = ttest_ind(group1, group2, equal_var=True)

print("Independent T-Test Results")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")


alpha = 0.05
print("\nInterpretation:")
print("Null Hypothesis: The means of both groups are equal.")
print("Alternative Hypothesis: The means of both groups are different.")
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel


np.random.seed(1)
before = np.random.normal(loc=70, scale=8, size=30)
after = before + np.random.normal(loc=-3, scale=5, size=30)

t_stat, p_val = ttest_rel(before, after)


print("Paired Sample T-test Results:")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.4f}")


alpha = 0.05
print("\nInterpretation:")
print("Null Hypothesis: No difference in means (before = after)")
print("Alternative Hypothesis: Means are different (before ≠ after)")

if p_val < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
print("\n")

plt.figure(figsize=(8, 5))
plt.plot(before, label="Before", marker='o')
plt.plot(after, label="After", marker='x')
plt.title("Before vs After Scores (Paired Sample)")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy.stats import norm, t

np.random.seed(42)
population = np.random.normal(loc=100, scale=15, size=10000)  # mean=100, std=15
pop_mean = 100
pop_std = 15


sample = np.random.choice(population, size=30, replace=False)
samp_mean = np.mean(sample)
samp_std = np.std(sample, ddof=1)
n = len(sample)

print(f"Sample Mean: {samp_mean:.2f}")
print(f"Sample Std Dev: {samp_std:.2f}")
print(f"Sample Size: {n}")


z_score = (samp_mean - pop_mean) / (pop_std / np.sqrt(n))
z_p_value = 2 * (1 - norm.cdf(abs(z_score)))


t_score = (samp_mean - pop_mean) / (samp_std / np.sqrt(n))
t_p_value = 2 * (1 - t.cdf(abs(t_score), df=n-1))


print("\nZ-Test Results:")
print(f"Z-Score: {z_score:.3f}, P-value: {z_p_value:.4f}")

print("\nT-Test Results:")
print(f"T-Score: {t_score:.3f}, P-value: {t_p_value:.4f}")


alpha = 0.05
print("\nConclusion at alpha = 0.05:")
if z_p_value < alpha:
    print("Z-test: Reject null hypothesis (significant difference)")
else:
    print("Z-test: Fail to reject null hypothesis")

if t_p_value < alpha:
    print("T-test: Reject null hypothesis (significant difference)")
else:
    print("T-test: Fail to reject null hypothesis")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)


    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return round(lower_bound, 3), round(upper_bound, 3), round(mean, 3)


np.random.seed(10)
sample_data = np.random.normal(loc=70, scale=10, size=30)
ci_lower, ci_upper, sample_mean = confidence_interval(sample_data, confidence=0.95)

print(f"Sample Mean = {sample_mean}")
print(f"95% Confidence Interval = ({ci_lower}, {ci_upper})")

print(f"This means we are 95% confident that the true population mean lies between {ci_lower} and {ci_upper}.")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_err = stats.sem(sample)


    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)


    moe = t_crit * std_err
    return round(moe, 3)

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=12, size=50)

moe_95 = margin_of_error(sample_data, confidence=0.95)
moe_99 = margin_of_error(sample_data, confidence=0.99)

print(f"Margin of Error (95% confidence): ±{moe_95}")
print(f"Margin of Error (99% confidence): ±{moe_99}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayesian_inference(p_disease, p_pos_given_disease, p_pos_given_no_disease):
    # P(No Disease)
    p_no_disease = 1 - p_disease

    # Total probability of a positive test
    p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

    # Bayes' Theorem
    p_disease_given_positive = (p_pos_given_disease * p_disease) / p_positive

    return round(p_disease_given_positive, 4)


p_disease = 0.01
p_pos_given_disease = 0.99
p_pos_given_no_disease = 0.05

posterior = bayesian_inference(p_disease, p_pos_given_disease, p_pos_given_no_disease)
print(f"Probability of having the disease given a positive test: {posterior * 100:.2f}%")
print("\nInterpretation")
print("Even if the test is 99% sensitive, if the disease is rare (1%), a positive test only gives a 16.67% chance of actually having the disease.")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

import pandas as pd
import seaborn as sns
from scipy.stats import chi2_contingency


df = sns.load_dataset('titanic')

contingency_table = pd.crosstab(df['survived'], df['pclass'])

print("Contingency Table:")
print(contingency_table)


chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print("\nChi-square Test Results:")
print(f"Chi-square Statistic = {chi2_stat:.3f}")
print(f"Degrees of Freedom = {dof}")
print(f"P-value = {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant association between Survived and Pclass.")
else:
    print("Fail to reject the null hypothesis: No significant association between Survived and Pclass.")
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

# Observed frequencies for each face of the die
observed = np.array([8, 9, 10, 11, 12, 10])

# Expected frequencies assuming a fair die (uniform distribution)
expected = np.array([10, 10, 10, 10, 10, 10])

# chi-square goodness-of-fit test
chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: The die is not fair.")
else:
    print("Fail to reject the null hypothesis: The die is fair.")
```

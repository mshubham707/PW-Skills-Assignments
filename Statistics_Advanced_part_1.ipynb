**Statistics Advanced Part 1**

1. What is a random variable in probability theory?

>> A random variable is a numerical outcome of a random phenomenon. It maps outcomes of a random process to real numbers.

2. What are the types of random variables?

>> There are two types: Discrete (countable outcomes) and Continuous (uncountable, measurable outcomes).

3. What is the difference between discrete and continuous distributions?

>> Discrete distributions have distinct values (like integers), while continuous distributions have outcomes over a continuous range.

4. What are probability distribution functions (PDF)?

>> A PDF describes the likelihood of a continuous random variable taking on a specific value (or range of values).

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

>> CDF gives the probability a variable is less than or equal to a value; PDF gives the probability density at a specific value.

6. What is a discrete uniform distribution?

>> A distribution where all outcomes are equally likely and finite, such as rolling a fair die.

7. What are the key properties of a Bernoulli distribution?

>> It has two outcomes (0 and 1), with a single trial and probability p of success.

8. What is the binomial distribution, and how is it used in probability?

>> It models the number of successes in n independent Bernoulli trials, each with success probability p.

9. What is the Poisson distribution and where is it applied?

>> It models the number of events occurring in a fixed interval of time/space, with a known constant mean rate.

10. What is a continuous uniform distribution?

>> A distribution where all values in a continuous range are equally likely, like picking a random number between 0 and 1.

11. What are the characteristics of a normal distribution?

>> It's symmetric, bell-shaped, centered at the mean, with most data within 3 standard deviations.

12. What is the standard normal distribution, and why is it important?

>> A normal distribution with mean 0 and standard deviation 1; it simplifies calculations in statistics.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

>> CLT states that the sampling distribution of the sample mean approaches normality as sample size increases, enabling inference.

14. How does the Central Limit Theorem relate to the normal distribution?

>> CLT explains why many distributions tend to be approximately normal when averaged over large samples.

15. What is the application of Z statistics in hypothesis testing?

>> Z-statistics are used to test hypotheses about population means when population variance is known.

16. How do you calculate a Z-score, and what does it represent?

>> Z = (X - μ) / σ; it represents how many standard deviations a value X is from the mean μ.

17. What are point estimates and interval estimates in statistics?

>> Point estimate gives a single value as an estimate, while interval estimate provides a range likely to contain the parameter.

18. What is the significance of confidence intervals in statistical analysis?

>> They give a range of values which is likely to contain the population parameter with a specified confidence level.

19. What is the relationship between a Z-score and a confidence interval?

>> Z-scores determine the margin of error in confidence intervals based on the desired confidence level.

20. How are Z-scores used to compare different distributions?

>> They standardize different distributions, allowing comparison across different units and scales.

21. What are the assumptions for applying the Central Limit Theorem?

>> Independent observations, identically distributed, and sufficiently large sample size (usually n > 30).

22. What is the concept of expected value in a probability distribution?

>> It's the long-term average outcome of a random variable: E(X) = Σ[x * P(x)].

23. How does a probability distribution relate to the expected outcome of a random variable?

>> The distribution defines the likelihood of outcomes, and the expected value summarizes their average effect.
